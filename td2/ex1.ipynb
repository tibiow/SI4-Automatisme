# Exercice 1

```python
#inporte numpy et le renomme en np
import numpy as np

#importe le sousmodule linalg de scipy
from scipy import linalg

#crée une matrice (2,2)
A = np.array([[1,2],[3,4]])

#crée une matrice (1,2)
b = np.array([[5],[6]])

#Résout l'équation Ax=b
x=np.linalg.solve(A,b)

#stock dans la variable r la résultat de Ax-b
r=A.dot(x)-b

#affiche r
print(r)
```

# Exercice 2

```python
#importe numpy et le renomme en np
import numpy as np

#importe matplotlib.pyplot et le renomme en plt
import matplotlib.pyplot as plt

#créer un vecteur x1 allant de 0 à 5 contenant 50 valeurs
x1 = np.linspace(0.0,5.0)

#créer un vecteur x2 allant de 0 à 2 contenant 50 valeurs
x2 = np.linspace(0.0,2.0)

#créer un vecteur y1 en fonction des valeurs de x1
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)

#créer un vecteur y2 en fonction des valeurs de x2
y2 = np.cos(2 * np.pi * x2)

#permet de découper la fenêtre graphique
plt.subplot(2,1,1)

#imprime un graphe en jaune(y) où tout les points sont des o(o) où tout les points sont reliés les uns aux autres(-)
#Ce graphe est crée à partir de x1 et y1 en associant 1 à 1 les valeurs de ces vecteurs
plt.plot(x1,y1,'yo-')

#permet de déterminer le nom du graphe
plt.title('A table of 2 subplots')

#permet de donner un nom à l'axe des ordonnées
plt.ylabel('Damped oscillation')

#permet de découper la fenêtre graphique
plt.subplot(2,1,2)

#imprime un graphe en rouge(r) où tout les points sont des .(.) où tout les points sont reliés les uns aux autres(-)
#Ce graphe est crée à partir de x2 et y2 en associant 1 à 1 les valeurs de ces vecteurs
plt.plot(x2,y2,'r.-')

#permet de donner un nom à l'axe des abscisses
plt.xlabel('time (s)')

#permet de donner un nom à l'axe des ordonnées 
plt.ylabel('undamped')

#permet de montrer les graphes
plt.show()
```

# Exercice 3

```python
#importe numpy et le renomme en np
import numpy as np

#importe le sousmodule linalg de scipy et le renomme en LA
from scipy import linalg as LA

#créer une matrice (3,3)
A = np.array([[2,0,0],[0,3,4],[0,4,9]])

#créer une matrice (1,3)
xk = np.array([[0.1],[0.1],[0.1]])

#permet de répeter 100 fois les instructions suivantes
for k in range(1,100):
    
    #on multiplie la matrice A par xk  et on entre la valeur dans yk
    yk=A.dot(xk)
    
    # on divise la matrice yk par la norme infinie de yk et on entre la valeur dans xk
    xk=yk/LA.norm(yk,np.inf)
    
#on divise la matrice xk par la norme 2 de xk et on entre la valeur dans x
x=xk/LA.norm(xk,2)

#on multiplie la transposé de x par A par x et on entre la valeur dans value
value = (x.transpose()).dot(A.dot(x))
```

Ce programme permet de déterminer la plus grande valeur propre d'une matrice

```python
#importe numpy et le renomme en np
import numpy as np

#importe matplotlib.pyplot et le renomme en plt
import matplotlib.pyplot as plt

#genère les 10 points aléatoirement
M=np.random.random((2,10))

#calcul le point moyen des 10 points
Mx=np.mean(M,1)

#affiche les 10 points en jaune avec des ronds
plt.plot(M[0,:],M[1,:],'yo')
#affiche le point moyen en rouge avec une croix
plt.plot(Mx[0],Mx[1],'r+')
#donne un titre à la courbe
plt.title("Random cloud and its center")
#label pour abscisse et ordonnée
plt.ylabel('ordonnée')
plt.xlabel('abscisse')
#print le graphe
plt.show()

```
