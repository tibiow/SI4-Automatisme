**2.3.1.** Import required libraries

```python
%matplotlib inline 
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model
import scipy
```

Generating the input vector

```python
n=500
x=np.linspace(1,100,n)
x=x.reshape(-1, 1) # to reshape properly the array (vertical vector)
```

**2.3.2.** Generate outputs following the given model (models for Exercises 2.2.10 and 2.2.11 can be defined here)

```python
noise=np.sqrt(104.04)*np.random.randn(n,1) # normal random noise
y=2*x+25.2+noise # linear model with noise and beta_1=2
# y=25.2+noise # linear model with noise and beta_1=0 (Exercise 2.2.10)
# y=2*x+0.02*x**2+25.2+noise # model with quadratic perturbation (Exercise 2.2.11)
```

```python
print(np.shape(x)) # verify the shape of each array
print(np.shape(noise))
print(np.shape(y))
```

**2.3.3.** Plot the cloud of points

```python
plt.scatter(x, y,  color='black')
plt.title('Cloud of points')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

**2.3.4.** Compute the regression parameters

```python
# Create the linear regression object
regr = linear_model.LinearRegression()
# Estimate the model using the data sets
regr.fit(x, y);
```

Print results related to the regression

```python
# The coefficients
print('Coefficients: %.2f' % regr.coef_)
print('Intercept: %.2f' % regr.intercept_)

# The mean square error
print('Residual variance estimate: %.2f'
      % (np.sum((regr.predict(x) - y) ** 2)/(n-2)))
```

**2.3.5.** Plot the regression line

```python
plt.scatter(x, y,  color='black')
plt.plot(x, regr.predict(x), color='blue',linewidth=3)
plt.title('Regression line')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

**2.3.6. and 2.3.7.** Estimate of $\sigma^2$ and residual analysis

```python
varhat=np.sum((regr.predict(x) - y) ** 2)/(n-2) # Estimate of the variance of the residuals
SSTx=np.sum((x - np.mean(x)) ** 2)
h=1/n+((x-np.mean(x))**2/SSTx)
residuals=(regr.predict(x) - y)
T=np.divide(residuals,np.sqrt(1-h))/np.sqrt(varhat) # standardized residuals
```

```python
plt.plot(x, T, color='red',linewidth=3)
plt.title('Standardized residuals')
plt.xlabel('x')
plt.ylabel('T')
plt.show()
```

**2.3.8.** Compute coefficient of determination

```python
print('Coefficient of determination: %.2f' % regr.score(x, y))
```

**2.3.9.** Significance test

```python
beta1hat=regr.coef_
t=beta1hat/np.sqrt(varhat/SSTx) # t-statistics
alpha=0.01 # significance level
thresholdQuantile = scipy.stats.t.ppf(1-alpha/2, n - 2) # threshold
```

```python
print('Decision statistics = %.2f' % t)
print('Decision threshold = %.2f' % thresholdQuantile)
```

As expected for the first simulated model, the coefficient of determination is close to one and we reject H0, meaning that the variation in $y$ is well explained by $x$ and that the linear model seems to be a good model.   

**2.3.10.** Simulated model with $\beta_1=0$ 

In this case we see that the coefficient of determination is close to zero and we accept H0, which means that there is no linear trend in the data (as expected).

**2.3.10.** Simulated model with second degree polynomial

Although $R^2\approx 1$ and rejection of H0 we can clealy see that there is still a quandratic trend in the residuals, meaning that the model may be nonlinear (which agrees with the simulated model).
