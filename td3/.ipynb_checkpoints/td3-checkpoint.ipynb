# Exercice 2

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
import scipy
%matplotlib inline

n=500

#créer les x
x=np.linspace(1,100,n)

x=x.reshape(-1,1)

sigma = np.sqrt(104.04)

#incertitude sur la mesure
e=sigma*np.random.random((n,1))

#vecteur de points possédant une incertitude de mesure
y = 2*x+25.2+e

#permet d'afficher le graphe de façon plus centré et plus lisible
plt.scatter(x,y)

#regression linéaire
regr = linear_model.LinearRegression()

regr.fit(x,y)

y_hat = regr.predict(x)

#permet de montrer la regression lineaire et les points de mesure réel.
plt.plot(x,y,'r.-')
plt.plot(x,y_hat,'y.-')


#calcul des residuals -> enlever les commentaires pour savoir le contenu du vecteur
residuals = y - y_hat
#print("residuals=")
#print(residuals)


#calcul des standdardized residuals (voir formule du cours) -> enlever les commentaires pour savoir le contenu du vecteur
sigma_xi_moyx_carre=0
for i in range(n):
    sigma_xi_moyx_carre+=(x[i]-(n/2))**2
hi=((1/n) + (((x-(n/2))**2)/sigma_xi_moyx_carre))
st_residuals = residuals /(sigma*np.sqrt(1-hi))
#print("st_residuals =")
#print(st_residuals)



#calcul du coefficient de détermination
sigma_yi_moy_y_carre=0
for i in range(n):
    sigma_yi_moy_y_carre+=(y[i]-(n/2))**2
sigma_yi_y_hat_carre=0
for i in range(n):
    sigma_yi_y_hat_carre+=(y[i]-y_hat[i])**2

deter=1-(sigma_yi_y_hat_carre/sigma_yi_moy_y_carre)
print("deter =")
print(deter)


#calcul de talpha2
# on décide d'une valeur pour alpha
alpha = 1
talpha2=scipy.stats.t.ppf(1-alpha/2,n-2)
print("talpha2 =")
print(talpha2)

plt.show()
```

Pour cette fonction les résidus sont compris entre -5 et +5 ce qui implique que notre précision reste assez bonne.

Le coefficient de détermination étant prêt de 1 on peut affirmer que la régression est proche de la réalité
