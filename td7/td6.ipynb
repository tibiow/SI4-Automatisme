# Question 5

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn import preprocessing
from sklearn import decomposition
import scipy
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline

file='turtles.csv'
df=pd.read_csv(file)

n=np.squeeze(np.shape(df['length']))
X=np.zeros((n,3))
X[:,0]=df['length']
X[:,1]=df['width']
X[:,2]=df['height']
label=df['sex']

X_scaled = preprocessing.scale(X)


#Q4
pca = decomposition.PCA(3).fit(X_scaled).explained_variance_
print("valeur des variances :")
print(pca)

#Q5

labels = ['PC1', 'PC2', 'PC3']
plt.plot([1, 2, 3], pca)
plt.plot([1, 2, 3], [pca[0],pca[0]+pca[1],pca[0]+pca[1]+pca[2]])
plt.xticks([1, 2, 3], labels)

plt.show()
```

Le graphe bleu reprsente les diffrentes variances et le graphe vert reprsente la somme cumules des variances.
