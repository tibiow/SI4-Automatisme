```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

%matplotlib inline

turtles = pd.read_csv('turtles.csv')

n = np.squeeze(np.shape(turtles['length']))
X = np.zeros((n, 3))
X[:, 0] = turtles['length']
X[:, 1] = turtles['width']
X[:, 2] = turtles['height']
label = turtles['sex']


def plot_turtles(data, sex, number):
    fig = plt.figure(figsize=(8, 6))
    dx = fig.add_subplot(111, projection='3d')
    dx.set_xlabel('Length')
    dx.set_ylabel('Width')
    dx.set_zlabel('Height')
    for i in range(0, number):
        if sex[i] == 'M':
            color_class = 'r'
            marker_class = '^'
        else:
            color_class = 'b'
            marker_class = 'o'
        dx.scatter(data[i, 0], data[i, 1], data[i, 2], c=color_class, marker=marker_class)


plot_turtles(X, label, n)
```

On voit que les males et les femelles ont l'air d'être séparés en deux groupes sur le graphe. Un ou plusieurs des 3 paramètres sont donc influencés par le sexe de la tortue.
