Q2) faire le calcul à la main

df/dx=0.2*x-8
->0.2*x-8=0 <=> x = 40


```python
import matplotlib.pyplot as plt
import numpy as np


def f_prime(a):
    y_prime=0.2*a-8
    return y_prime

def f(a):
    y_=0.1*a**2-8*a+10
    return y_



N = 20
x=np.linspace(0,N,N)
x0=-50
step = 2

xk=[x0]
f_prime_xk=[]
epsilon=[]
x_star=40

for i in range(N):
    #calcul f_prime_xk
    f_prime_xk.append(f_prime(xk[i]))
    #calcul epsilon
    epsilon.append(f(xk[i])-f(x_star))
    #calcul xk 
    xk.append(xk[i]-step*f_prime_xk[i]) 
    
#Q4
plt.subplot(221)
plt.title("epsilon en fonction de k")
plt.plot(x,epsilon,'r.-')
plt.subplot(222)
plt.title("f_prime_xk en fonction de k")
plt.plot(x,f_prime_xk,'r.-')

plt.subplot(212)
##display de la courbe de la première question
x1=np.linspace(-60,140,1000)
x1=x1.reshape(-1,1)
y=0.1*x1**2-8*x1+10
plt.scatter(x1,y)
plt.plot(x1,y,'r.-')
##display des différentes échantillons effectuées avec le gradient algorithm
xk.pop(N)
yk=[]
for i in range(N):
    yk.append(f(xk[i]))
plt.plot(xk,yk,'yo')


plt.show()
```
